```python
import torch
```

```python
x = torch.tensor(3.0,requires_grad=True)
```

```python
y = x**2
y
```

```python
z = torch.sin(y)
z
```

```python
z.backward()
```

```python
x.grad
```

```python
b = torch.mean(a**2)
b
```

```python
b.backward()
```

```python
print(a.grad)
```
